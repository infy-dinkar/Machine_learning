```python
### Idea: "If a passenger's age is missing, fill it with the average age of other passengers in the same class. Otherwise, keep the original age."

# Step 1: Compute mean age for each Pclass
class_avg_age = df.groupby('Pclass')['Age'].mean()

# Step 2: Custom function to fill missing Age values
def fill_age(row):
    if pd.isnull(row['Age']):
        return class_avg_age[row['Pclass']]
    else:
        return row['Age']

# Step 3: Apply function to the dataset
df['Age'] = df.apply(fill_age, axis=1)

#we just have filled nul values in age according to the class Average
```

```python
X = df.drop('Survived', axis = 1)
y = df['Survived']
```

```python
# Importing required method from sklearn
from sklearn.model_selection import train_test_split
```

```python
# Let's keep the default size and states
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logR = LogisticRegression()
```

```python
# Training the model on our training dataset -- shift+tab for doc string
logR.fit(X_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_train=X_train_scaled
X_test_scaled = scaler.transform(X_test)
X_test=X_test_scaled
logR.fit(X_train, y_train)
```

```python
predictions = logR.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
# Let's pass the y_test and predictions to get the confusion_matrix
print(confusion_matrix(y_test, predictions))
```

```python
# Get confusion matrix values
cm = confusion_matrix(y_test, predictions)

# Plot it with seaborn heatmap
plt.figure(figsize=(5,3))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Survived', 'Survived'], yticklabels=['Not Survived', 'Survived'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
